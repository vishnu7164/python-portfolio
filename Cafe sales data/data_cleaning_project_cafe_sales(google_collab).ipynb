### Assumption is that the price per unit item does not change

Viewing the above table it is clear we can extract the price of items and map corresponing price to replace NaN values in 'Price Per Unit' column

```python
unique_items = [item for item in df['Item'].unique() if item is not np.nan]
unique_items
```

```python
item_price_dict = {}

for item in unique_items:
  item_price_dict[item] = df[df['Item']== item]['Price Per Unit'].value_counts().index

item_price_dict
```

```python
# cleaning up item_price_dict

for key in item_price_dict:
  item_price_dict[key] = item_price_dict[key][0]

item_price_dict
```

Unique value of items are extracted from the column and respective prices are saved into a dictonary of key value pairs for the purpose of mapping with 'Price Per Unit' column

```python
df['Price Per Unit'] = df['Price Per Unit'].fillna(df['Item'].map(item_price_dict))
```

```python
df['Price Per Unit'].isna().sum()
```

Assuming same logic as price of item doesn't for an item and vice versa

#### Reversing the key value pairs in dictionary to get the item value with respect to price.


```python
price_item = { price:item for item,price in item_price_dict.items()}
price_item
```

```python
df['Item'] = df['Item'].fillna(df['Price Per Unit'].map(price_item))
```
